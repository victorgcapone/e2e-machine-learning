```python
import pandas as pd 
```

```python
data = pd.read_csv("../datasets/titanic.tsv", sep="\t")
data.head()
```

```python
#Let's check for missing values
data.isna().sum()
```

```python
# We will also assume that each missing Age is the average age
data.Age.fillna(data.Age.mean(), inplace=True)
data.isna().sum()
```

```python
# with that many missing values, we can drop the cabin Columns, 
# let's also drop that single NaN in the embarked column
# We will alos drop the name columns
data = data.dropna()
data = data.drop(["Name", "Cabin"], axis=1)
data.isna().sum()
```

```python
# For our purposes the ticket doesn't matter, as well as the passenger Id
data = data.drop(['Ticket', 'PassengerId'], axis=1)
data.head()
```

```python
# Let's One-Hot-Encode our Embarked and Sex columns using pandas get_dummies
data = pd.get_dummies(data, columns=["Embarked", "Sex"])
```
